```python
# ============================================================
# PRiFAREDâ€“MADDPG (Retry-Aware, Bounded Reward) â€” Colab Notebook
# End-to-end, single-cell, runnable (500 episodes; 5 agents by default)
# ============================================================

import os, math, random, sys, time, json
import numpy as np
import pandas as pd
from collections import deque, namedtuple
from typing import Dict, List, Tuple
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim

# ----------------- Repro -----------------
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)

# ----------------- Config -----------------
class CFG:
    # Agents / federation scale
    N_AGENTS = 5                 # change to 7 for larger federation
    TOP_K_TASKS = 8              # tasks each agent "sees" per step
    MMC_SERVERS = 3              # M/M/c queue: c servers per ES
    R_MAX = 3                    # retry cap for normalization

    # Reward weights (bounded design)
    ALPHA_T = 0.55               # weight for latency (normalized)
    BETA_E  = 0.25               # weight for energy (normalized)
    DELTA_U = 0.15               # weight for unfairness (1 - J)
    GAMMA_R = 0.20               # bonus for success-after-retry

    # Nets / dims
    OBS_DIM = 6                  # [queue_proxy, Frem, Brem, Prem, Rbar, Dbar]
    ACT_DIM = 4                  # [xcpu, xbw, xpow, xoff] (raw; projected)
    HIDDEN  = 192
    ACTOR_LR = 1e-4
    CRITIC_LR= 1e-3
    GAMMA    = 0.98
    TAU      = 0.005

    # Replay / training
    BUFFER_SIZE   = 120_000
    BATCH_SIZE    = 512
    EPISODES      = 500          # <- per request
    STEPS_PER_EP  = 220
    TRAIN_AFTER   = 2_000
    TRAIN_EVERY   = 40
    UPDATES_PER   = 1

    # Exploration (Gaussian) noise schedule
    NOISE_STD_INIT = 0.25
    NOISE_STD_FINAL= 0.04
    NOISE_DECAY_STEPS = 60_000

    # Split
    TRAIN_ROWS = 4000
    TEST_ROWS  = 1000

    # Normalizers (energy target for scale; tweak if your dataset differs)
    E_MAX = 2.0                  # J (scale for normalization; updated online too)
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Device:", CFG.DEVICE)

# ============================================================
# 0) Load CSV (or synthesize a 5k dataset with same schema)
# ============================================================
try:
    from google.colab import files
    print("ðŸ“¥ Upload your CSV (e.g., priafred_maddpg_dataset_5000.csv).")
    uploaded = files.upload()
except Exception:
    uploaded = {}

if uploaded:
    fname = list(uploaded.keys())[0]
    df_full = pd.read_csv(fname)
else:
    print("No upload detected â€” generating synthetic 5k dataset matching required columnsâ€¦")
    NROWS = CFG.TRAIN_ROWS + CFG.TEST_ROWS
    rows = []
    for tid in range(NROWS):
        rows.append({
            'Task_ID': tid+1,
            'C_cpu(GC)': np.random.uniform(10, 80),      # Giga-cycles
            'C_data(Mbits)': np.random.uniform(1, 80),   # Mbits
            'T_max(ms)': np.random.uniform(100, 2000),   # ms
            'Rn(Retry)': np.random.randint(0, 4),
            'Battery Level(%)': np.random.uniform(10, 100),
            'Edge_Server_ID': np.random.randint(1, CFG.N_AGENTS+1),
            'F_max(GHz)': np.random.uniform(2.5, 5.0),
            'B_max(MHz)': np.random.uniform(5, 30),
            'P_max(W)': np.random.uniform(2.5, 6.0),
            'Service_Class': np.random.choice([0,1,2]),  # 0=N,1=M,2=U
            'Premium_User': np.random.choice([0,1]),
            'Channel_SNR(dB)': np.random.uniform(5, 25),
        })
    df_full = pd.DataFrame(rows)

print("Dataset shape:", df_full.shape)
display(df_full.head(3))

# Column map (edit here to match your file headers if needed)
COLUMN_MAP = {
    'task_id':     'Task ID',
    'cycles_GC':   'C_cpu (GC) [10-100]',
    'data_Mbits':  'C_data (Mbits) [5-50]',
    'tmax_ms':     'T_max (ms) [100-500]',
    'retry':       'Rn (Retry)',
    'battery_pct': 'Battery Level(%)',
    'server_id':   'Edge_Server_ID',
    'Fmax_GHz':    'F_max(GHz)',
    'Bmax_MHz':    'B_max(MHz)',
    'Pmax_W':      'P_max(W)',
    'svc_class':   'Service_Class',
    'premium':     'Premium_User',
    'snr_db':      'Channel_SNR(dB)',
}

# Sanity: ensure columns exist (create defaults if missing)
def ensure_col(df, name, default):
    if name not in df.columns:
        df[name] = default

for k,v in COLUMN_MAP.items():
    if v not in df_full.columns:
        if k in ['svc_class','premium','retry','server_id']:
            ensure_col(df_full, v, 0)
        elif k=='snr_db':
            ensure_col(df_full, v, 15.0)
        elif k=='Fmax_GHz':
            ensure_col(df_full, v, 4.0)
        elif k=='Bmax_MHz':
            ensure_col(df_full, v, 20.0)
        elif k=='Pmax_W':
            ensure_col(df_full, v, 5.0)
        elif k=='battery_pct':
            ensure_col(df_full, v, 50.0)
        elif k=='tmax_ms':
            ensure_col(df_full, v, 1000.0)

# Shuffle & split
df = df_full.sample(frac=1.0, random_state=SEED).reset_index(drop=True)
df_train = df.iloc[:CFG.TRAIN_ROWS].reset_index(drop=True)
df_test  = df.iloc[CFG.TRAIN_ROWS:CFG.TRAIN_ROWS+CFG.TEST_ROWS].reset_index(drop=True)
print("Train:", df_train.shape, "Test:", df_test.shape)

# ============================================================
# 1) Utility: Priority + Shannon + Queue + Fairness
# ============================================================
def service_scalar(svc_class:int)->float:
    return [0.33, 0.67, 1.0][int(np.clip(svc_class,0,2))]

def deadline_sensitivity(tmax_ms:float, df_all:pd.DataFrame)->float:
    invT = 1.0 / (tmax_ms/1000.0 + 1e-9)
    max_invT = 1.0 / (df_all[COLUMN_MAP['tmax_ms']].max()/1000.0 + 1e-9)
    return float(np.clip(invT / (max_invT + 1e-9), 0, 1))

def battery_constraint(batt_pct:float)->float:
    return float(np.clip(1.0 - batt_pct/100.0, 0.0, 1.0))

def retry_factor(retry:int)->float:
    return float(min(max(retry,0)/CFG.R_MAX, 1.0))

def compute_priority(row, df_all,
                     alpha_base={'D':0.30,'S':0.30,'th':0.20,'eta':0.20},
                     boosts={'D':0.20,'S':0.20,'th':0.20,'eta':0.20},
                     beta=0.2) -> Tuple[float, Dict]:
    D  = deadline_sensitivity(row[COLUMN_MAP['tmax_ms']], df_all)
    S  = service_scalar(row[COLUMN_MAP['svc_class']])
    th = battery_constraint(row[COLUMN_MAP['battery_pct']])
    eta= retry_factor(row[COLUMN_MAP['retry']])

    cD   = (row[COLUMN_MAP['tmax_ms']] < 300)
    cS   = (S >= 0.999)
    cth  = (row[COLUMN_MAP['battery_pct']] < 15.0)
    ceta = (row[COLUMN_MAP['retry']] >= CFG.R_MAX)

    aD   = alpha_base['D']  + (boosts['D']   if cD   else 0.0)
    aS   = alpha_base['S']  + (boosts['S']   if cS   else 0.0)
    ath  = alpha_base['th'] + (boosts['th']  if cth  else 0.0)
    aeta = alpha_base['eta']+ (boosts['eta'] if ceta else 0.0)
    s = aD + aS + ath + aeta + 1e-9
    aD,aS,ath,aeta = aD/s, aS/s, ath/s, aeta/s
    phi = aD*D + aS*S + ath*th + aeta*eta
    W = (1.0 - beta)*phi + beta*float(row.get(COLUMN_MAP['premium'],0))
    return float(np.clip(W,0,1)), {'D':D,'S':S,'th':th,'eta':eta}

def shannon_rate(B_MHz:float, SNR_dB:float)->float:
    B = max(B_MHz, 1e-6) * 1e6
    snr_lin = 10**(SNR_dB/10.0)
    return B * math.log2(1.0 + snr_lin)  # bits/s

def mmc_waiting_time(lmbda:float, mu:float, c:int)->float:
    if c <= 0 or mu <= 0: return 0.0
    rho = lmbda/(c*mu + 1e-12)
    if rho >= 1.0:  # unstable -> large queue; cap
        return 1.0
    summ = sum([(lmbda/mu)**k / math.factorial(k) for k in range(c)])
    last = ((lmbda/mu)**c / math.factorial(c)) * (1.0/(1.0 - rho))
    P0 = 1.0 / (summ + last + 1e-12)
    Pwait = ((lmbda/mu)**c / math.factorial(c)) * (1.0/(1.0 - rho)) * P0
    Wq = Pwait * (1.0/(c*mu - lmbda + 1e-12))
    return max(Wq, 0.0)

def jains_index(loads:List[float])->float:
    x = np.array(loads) + 1e-9
    return (x.sum()**2) / (len(x) * (x**2).sum())

# ============================================================
# 2) Action projection and mapping
# ============================================================
def project_action_decode(raw_action:np.ndarray, caps:Dict)->Dict:
    # raw_action: [x_cpu, x_bw, x_pow, x_off] âˆˆ R
    x = np.maximum(raw_action[:3], 0.0)
    s = x.sum()
    hat = np.array([1/3,1/3,1/3]) if s < 1e-9 else x/s
    f_alloc = hat[0] * caps['Fmax']                # Hz
    b_alloc = hat[1] * caps['Bmax'] * 1e6         # MHz->Hz
    p_alloc = hat[2] * caps['Pmax']               # W
    x_off   = 1/(1+np.exp(-raw_action[3]))        # sigmoid in [0,1]
    return {'f_alloc': f_alloc, 'b_alloc': b_alloc, 'p_alloc': p_alloc, 'x_off': x_off}

# ============================================================
# 3) Environment: Federated MEC (multi-agent)
# ============================================================
class FederatedMECEnv:
    def __init__(self, df_all:pd.DataFrame, df_for_norm:pd.DataFrame, n_agents:int=CFG.N_AGENTS):
        self.df = df_all.reset_index(drop=True)
        self.df_norm = df_for_norm
        self.n = n_agents
        self.ptr = 0
        self.max_idx = len(self.df)-1
        self.loads = np.zeros(self.n, dtype=float)     # for Jain
        self.E_max_track = CFG.E_MAX                   # online energy scaler

    def reset(self):
        self.ptr = 0
        self.loads[:] = 0.0
        return self._obs_all()

    def _obs_agent(self, agent_id:int, rows:pd.DataFrame):
        df_a = rows[rows[COLUMN_MAP['server_id']] == (agent_id+1)]
        if len(df_a)==0: df_a = rows.sample(min(len(rows), CFG.TOP_K_TASKS))
        top = df_a.head(CFG.TOP_K_TASKS)

        Rbar = float(np.mean(top[COLUMN_MAP['retry']])) if len(top)>0 else 0.0
        Dbar = float(np.mean([deadline_sensitivity(v, self.df_norm) for v in top[COLUMN_MAP['tmax_ms']]])) if len(top)>0 else 0.0

        Fmax = float(top[COLUMN_MAP['Fmax_GHz']].mean() if len(top)>0 else 4.0) * 1e9
        Bmax = float(top[COLUMN_MAP['Bmax_MHz']].mean() if len(top)>0 else 20.0)
        Pmax = float(top[COLUMN_MAP['Pmax_W']].mean() if len(top)>0 else 5.0)

        obs = np.array([
            min(1.0, len(top)/CFG.TOP_K_TASKS),   # queue proxy
            Fmax/6e9,                              # normalize
            Bmax/30.0,
            Pmax/10.0,
            Rbar/CFG.R_MAX,
            Dbar
        ], dtype=np.float32)

        cap = {'Fmax':Fmax, 'Bmax':Bmax, 'Pmax':Pmax}
        tasks = top
        return obs, cap, tasks

    def _obs_all(self):
        i0, i1 = self.ptr, min(self.ptr + CFG.TOP_K_TASKS*self.n, self.max_idx+1)
        rows = self.df.iloc[i0:i1]
        obs_list, caps, tasks = [], [], []
        for a in range(self.n):
            o, cap, tks = self._obs_agent(a, rows)
            obs_list.append(o); caps.append(cap); tasks.append(tks)
        return obs_list, caps, tasks

    def _compute_latency_energy(self, row, alloc:Dict, off_ratio:float):
        # Data & cycles split
        data_bits   = float(row[COLUMN_MAP['data_Mbits']]) * 1e6
        tmax_s      = float(row[COLUMN_MAP['tmax_ms']]) / 1000.0
        cycles_total= float(row[COLUMN_MAP['cycles_GC']]) * 1e9

        data_off   = off_ratio * data_bits
        data_local = data_bits - data_off
        cycles_local = cycles_total * (data_local/(data_bits+1e-9))
        cycles_off   = cycles_total - cycles_local

        # Transmission
        rate = shannon_rate(float(row[COLUMN_MAP['Bmax_MHz']]), float(row[COLUMN_MAP['snr_db']]))
        T_tx = data_off / (rate + 1e-9)

        # Compute (local+edge) â€” simple edge factor
        f = alloc['f_alloc']
        T_comp_local = (cycles_local/(f + 1e-9)) if f>1e-6 else 1.0
        T_comp_edge  = (cycles_off /(1.5*f + 1e-9)) if (data_off>0 and f>1e-6) else 0.0

        # Queue (M/M/c)
        mean_service = max(1e-6, (T_comp_local + T_comp_edge))
        mu = 1.0/mean_service
        lmbda = 0.6 * mu
        T_queue = mmc_waiting_time(lmbda, mu, CFG.MMC_SERVERS)

        # Total latency
        T_total = T_tx + T_comp_local + T_comp_edge + T_queue
        T_total = min(T_total, 3.0*tmax_s)  # cap to 3x deadline

        # Energy (DVFS + radio)
        kappa = 1e-27
        E_comp = kappa * (f**2) * (T_comp_local + T_comp_edge)
        E_tx   = alloc['p_alloc'] * T_tx
        E_total= E_comp + E_tx

        # Track dynamic energy scale
        self.E_max_track = max(self.E_max_track, E_total + 1e-9)

        success = 1 if T_total <= tmax_s else 0
        return T_total, E_total, success

    def step(self, actions:List[np.ndarray]):
        (obs_list, caps, tasks) = self._obs_all()
        rewards = []
        lats, engs, succs = [], [], []

        for a_id in range(self.n):
            cap = caps[a_id]; tks = tasks[a_id]
            raw = actions[a_id]
            alloc = project_action_decode(raw, cap)

            if len(tks)==0:
                rewards.append(0.0); lats.append(0.0); engs.append(0.0); succs.append(0.0); continue

            # Priority-based split for offload share
            Wvec = []
            for _, row in tks.iterrows():
                W,_ = compute_priority(row, self.df_norm)
                Wvec.append(W)
            Wvec = np.array(Wvec); Wvec = Wvec/(Wvec.sum()+1e-9)

            lat_sum, eng_sum, succ_sum = 0.0, 0.0, 0.0
            for idx, (_, row) in enumerate(tks.iterrows()):
                off_ratio = float(Wvec[idx]) * alloc['x_off']
                Lt, Et, Sc = self._compute_latency_energy(row, alloc, off_ratio)
                lat_sum += Lt; eng_sum += Et; succ_sum += Sc

            # Update load proxy for fairness
            self.loads[a_id] = 0.6*self.loads[a_id] + 0.4*(alloc['f_alloc']/(cap['Fmax']+1e-9))

            # Bounded reward in [-1,1]
            Tmax = np.mean(tks[COLUMN_MAP['tmax_ms']].values)/1000.0
            Tnorm = (lat_sum/len(tks)) / (Tmax + 1e-9)
            Enorm = (eng_sum/len(tks)) / (self.E_max_track + 1e-9)
            Tnorm = np.clip(Tnorm, 0.0, 1.0)  # bounded
            Enorm = np.clip(Enorm, 0.0, 1.0)

            r = 1.0 - (CFG.ALPHA_T*Tnorm + CFG.BETA_E*Enorm)  # base
            # retry-aware bonus: success for any task with retry>0
            if (tks[COLUMN_MAP['retry']].values > 0).any() and succ_sum>0:
                r += CFG.GAMMA_R * (succ_sum/len(tks))
            rewards.append(float(np.clip(r, -1.0, 1.0)))

            lats.append(lat_sum); engs.append(eng_sum); succs.append(succ_sum)

        # Fairness penalty
        J = jains_index(self.loads)
        rewards = [float(np.clip(rv - CFG.DELTA_U*max(0.0, 1.0 - J), -1.0, 1.0)) for rv in rewards]

        # Advance sliding window
        self.ptr = min(self.ptr + CFG.TOP_K_TASKS*self.n, self.max_idx+1)
        done = (self.ptr >= self.max_idx)
        next_obs = self._obs_all()[0]
        info = {'fairness': J, 'lat': float(np.mean(lats)), 'eng': float(np.mean(engs)), 'succ': float(np.mean(succs))}
        return next_obs, rewards, done, info

# ============================================================
# 4) Replay Buffer
# ============================================================
Transition = namedtuple('Transition', ('s','o_list','a','r','s2','o2_list','done'))
class ReplayBuffer:
    def __init__(self, cap=CFG.BUFFER_SIZE):
        self.buf = deque(maxlen=cap)
    def push(self, *args): self.buf.append(Transition(*args))
    def sample(self, batch_size):
        idxs = np.random.choice(len(self.buf), batch_size, replace=False)
        return [self.buf[i] for i in idxs]
    def __len__(self): return len(self.buf)

# ============================================================
# 5) Networks
# ============================================================
class MLP(nn.Module):
    def __init__(self, inp, out, hid=CFG.HIDDEN, out_act='tanh'):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(inp, hid), nn.ReLU(),
            nn.Linear(hid, hid), nn.ReLU(),
            nn.Linear(hid, out)
        )
        self.out_act = out_act
    def forward(self, x):
        y = self.net(x)
        if self.out_act=='tanh': return torch.tanh(y)
        return y

class MADDPG:
    def __init__(self, n_agents:int, obs_dim:int, act_dim:int, device=CFG.DEVICE):
        self.n = n_agents; self.obs_dim=obs_dim; self.act_dim=act_dim
        self.device = device
        self.actors   = [MLP(obs_dim, act_dim).to(device) for _ in range(self.n)]
        self.critics  = [MLP(self.n*(obs_dim+act_dim), 1, hid=256, out_act=None).to(device) for _ in range(self.n)]
        self.actors_t = [MLP(obs_dim, act_dim).to(device) for _ in range(self.n)]
        self.critics_t= [MLP(self.n*(obs_dim+act_dim), 1, hid=256, out_act=None).to(device) for _ in range(self.n)]
        for i in range(self.n):
            self.actors_t[i].load_state_dict(self.actors[i].state_dict())
            self.critics_t[i].load_state_dict(self.critics[i].state_dict())
        self.optA = [optim.Adam(self.actors[i].parameters(), lr=CFG.ACTOR_LR) for i in range(self.n)]
        self.optC = [optim.Adam(self.critics[i].parameters(), lr=CFG.CRITIC_LR) for i in range(self.n)]
        self.rb = ReplayBuffer()

    def select_actions(self, obs_list, noise_std=0.1, eval_mode=False):
        acts=[]
        for i in range(self.n):
            o = torch.tensor(obs_list[i], dtype=torch.float32, device=CFG.DEVICE).unsqueeze(0)
            a = self.actors[i](o).detach().cpu().numpy()[0]
            if not eval_mode and noise_std>0:
                a += np.random.normal(0, noise_std, size=a.shape)
            acts.append(a)
        return acts

    def update(self):
        if len(self.rb) < CFG.BATCH_SIZE: return
        batch = self.rb.sample(CFG.BATCH_SIZE)

        o1 = np.array([np.concatenate(t.o_list, axis=0) for t in batch], dtype=np.float32)
        o2 = np.array([np.concatenate(t.o2_list, axis=0) for t in batch], dtype=np.float32)
        a1 = np.array([np.concatenate(t.a, axis=0) for t in batch], dtype=np.float32)
        r  = np.array([np.array(t.r, dtype=np.float32) for t in batch], dtype=np.float32)
        d  = np.array([t.done for t in batch], dtype=np.float32)

        o1 = torch.tensor(o1, device=CFG.DEVICE)
        o2 = torch.tensor(o2, device=CFG.DEVICE)
        a1 = torch.tensor(a1, device=CFG.DEVICE)
        r  = torch.tensor(r,  device=CFG.DEVICE)
        d  = torch.tensor(d,  device=CFG.DEVICE).unsqueeze(-1)

        # next actions (targets)
        with torch.no_grad():
            split = self.obs_dim
            o2_parts = [o2[:, i*split:(i+1)*split] for i in range(self.n)]
            a2_parts = [self.actors_t[i](o2_parts[i]) for i in range(self.n)]
            a2 = torch.cat(a2_parts, dim=-1)

        # critic updates
        for i in range(self.n):
            x1 = torch.cat([o1, a1], dim=-1)
            with torch.no_grad():
                x2 = torch.cat([o2, a2], dim=-1)
                q_tgt = self.critics_t[i](x2)
                y = r[:, i:i+1] + CFG.GAMMA*(1.0 - d)*q_tgt
            q = self.critics[i](x1)
            loss_c = ((q - y)**2).mean()
            self.optC[i].zero_grad(); loss_c.backward(); self.optC[i].step()

        # actor updates
        for i in range(self.n):
            split = self.obs_dim
            o1_parts = [o1[:, j*split:(j+1)*split] for j in range(self.n)]
            a_pred_parts = [self.actors[j](o1_parts[j]) for j in range(self.n)]
            a_pred = torch.cat(a_pred_parts, dim=-1)
            x = torch.cat([o1, a_pred], dim=-1)
            loss_a = - self.critics[i](x).mean()
            self.optA[i].zero_grad(); loss_a.backward(); self.optA[i].step()

        # soft targets
        for i in range(self.n):
            with torch.no_grad():
                for t,s in zip(self.actors_t[i].parameters(), self.actors[i].parameters()):
                    t.data.mul_(1-CFG.TAU); t.data.add_(CFG.TAU*s.data)
                for t,s in zip(self.critics_t[i].parameters(), self.critics[i].parameters()):
                    t.data.mul_(1-CFG.TAU); t.data.add_(CFG.TAU*s.data)

    def push(self, s, o_list, a, r, s2, o2_list, done):
        self.rb.push(s, o_list, a, r, s2, o2_list, done)

# ============================================================
# 6) Build envs
# ============================================================
env_train = FederatedMECEnv(df_train, df_train, n_agents=CFG.N_AGENTS)
env_test  = FederatedMECEnv(df_test,  df_train, n_agents=CFG.N_AGENTS)  # normalize w/ train stats
maddpg = MADDPG(CFG.N_AGENTS, CFG.OBS_DIM, CFG.ACT_DIM)

# ============================================================
# 7) Training loop (500 episodes)
# ============================================================
def noise_schedule(step:int)->float:
    return float(np.interp(step,
                           [0, CFG.NOISE_DECAY_STEPS],
                           [CFG.NOISE_STD_INIT, CFG.NOISE_STD_FINAL]))

reward_hist, latency_hist, energy_hist, fairness_hist = [], [], [], []
global_step = 0
best_eval = None

for ep in range(CFG.EPISODES):
    obs_list, caps, tasks = env_train.reset()
    ep_rew = 0.0; ep_lat=0.0; ep_eng=0.0; ep_J=0.0; steps=0

    for t in range(CFG.STEPS_PER_EP):
        actions = maddpg.select_actions(obs_list, noise_std=noise_schedule(global_step), eval_mode=False)
        next_obs_list, rewards, done, info = env_train.step(actions)

        s  = np.concatenate(obs_list, axis=0)
        s2 = np.concatenate(next_obs_list, axis=0)
        maddpg.push(s, obs_list, actions, rewards, s2, next_obs_list, float(done))

        obs_list = next_obs_list
        ep_rew += np.mean(rewards); ep_lat += info['lat']; ep_eng += info['eng']; ep_J += info['fairness']; steps+=1
        global_step += 1

        if (len(maddpg.rb) > CFG.TRAIN_AFTER) and (global_step % CFG.TRAIN_EVERY == 0):
            for _ in range(CFG.UPDATES_PER):
                maddpg.update()

        if done: break

    reward_hist.append(ep_rew/steps)
    latency_hist.append(ep_lat/steps)
    energy_hist.append(ep_eng/steps)
    fairness_hist.append(ep_J/steps)

    if (ep+1) % 10 == 0:
        print(f"[Train] Ep {ep+1}/{CFG.EPISODES} | R(avg): {reward_hist[-1]:.3f} | "
              f"J: {fairness_hist[-1]:.3f} | Lat(avg): {latency_hist[-1]:.3f}s | Eng(avg): {energy_hist[-1]:.3e}J")

    # periodic eval
    if (ep+1) % 50 == 0:
        def evaluate(env, maddpg, episodes=4):
            mets={'reward':[], 'lat':[], 'eng':[], 'J':[]}
            for _ in range(episodes):
                obs_list, caps, tasks = env.reset()
                epR=0.0; it=0
                while True:
                    actions = maddpg.select_actions(obs_list, noise_std=0.0, eval_mode=True)
                    next_obs_list, rewards, done, info = env.step(actions)
                    obs_list = next_obs_list
                    epR += np.mean(rewards); it+=1
                    if done: break
                mets['reward'].append(epR/it); mets['lat'].append(info['lat']); mets['eng'].append(info['eng']); mets['J'].append(info['fairness'])
            return {k: float(np.mean(v)) for k,v in mets.items()}
        ev = evaluate(env_test, maddpg, episodes=4)
        print(f"[Eval] Ep {ep+1}: R={ev['reward']:.3f} | J={ev['J']:.3f} | Lat={ev['lat']:.3f}s | Eng={ev['eng']:.3e}J")
        # save best
        score = ev['reward'] + ev['J'] - 0.5*ev['lat']
        if (best_eval is None) or (score > best_eval[0]):
            best_eval = (score, ev)
            for i in range(CFG.N_AGENTS):
                torch.save(maddpg.actors[i].state_dict(), f"actor_agent_{i}.pt")
                torch.save(maddpg.critics[i].state_dict(), f"critic_agent_{i}.pt")
            print("âœ… Saved best actor/critic weights.")

# ============================================================
# 8) Plots
# ============================================================
def moving_avg(x, k=10):
    x = np.array(x, dtype=float);
    if len(x)<k: return x
    return np.convolve(x, np.ones(k)/k, mode='valid')

plt.figure(figsize=(14,4))
plt.subplot(1,4,1); plt.plot(moving_avg(reward_hist)); plt.title("Train: Avg Reward (MA)")
plt.subplot(1,4,2); plt.plot(moving_avg(latency_hist)); plt.title("Train: Avg Latency (s, MA)")
plt.subplot(1,4,3); plt.plot(moving_avg(energy_hist)); plt.title("Train: Avg Energy (J, MA)")
plt.subplot(1,4,4); plt.plot(moving_avg(fairness_hist)); plt.title("Train: Jain's Fairness (MA)")
plt.tight_layout(); plt.show()

print("\nBest Eval (if saved):"), best_eval[1] if best_eval else "N/A"
```
